### Data Cleaning 

```python
df.isnull().sum()
```

```python
# Convert everything to string first
df['Mileage'] = df['Mileage'].astype(str)

# Remove both ' kmpl' and ' km/kg'
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '', regex=False)
df['Mileage'] = df['Mileage'].str.replace(' km/kg', '', regex=False)

# Convert to numeric (NaN if invalid)
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')
```

```python
df['Engine'] = df['Engine'].astype(str).str.replace(' CC', '', regex=False)
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')

df['Power'] = df['Power'].astype(str).str.replace(' bhp', '', regex=False)
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')
```

```python
df
```

### Feature Engineering

```python
# Brand Extraction:

df['brand'] = df['Name'].str.split().str[0]
```

```python
# Encode categoroical columns:

df = pd.get_dummies(df, columns=['Fuel_Type','Transmission','Owner_Type'], drop_first=True)
```

### Model Preparation

```python
df
```

```python
df = df.dropna(subset=['Price'])
```

```python
features = ['Year','Kilometers_Driven','Mileage','Engine','Power','Seats','brand',
           'Fuel_Type_Diesel', 'Fuel_Type_Electric', 'Fuel_Type_LPG', 'Fuel_Type_Petrol',
    'Transmission_Manual',
    'Owner_Type_Fourth & Above', 'Owner_Type_Second', 'Owner_Type_Third']
```

```python
x = df[features]
y = df['Price']
```

```python
if x['brand'].dtype == 'object':
    x = pd.get_dummies(x, columns=['brand'], drop_first=True)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_test
```

```python
y_train
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
